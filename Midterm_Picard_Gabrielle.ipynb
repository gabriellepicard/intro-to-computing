```python
for element in rand_list:
    if element in dictionary:
        dictionary[element] += 1
    else:
        dictionary[element] = 1
        
for key, value in dictionary.items():
    print("% d: %d"%(key,value))
```

```python
# Step 1: import pandas, numpy & matplotlib already imported
import pandas as pd
```

```python
# Preview
df = pd.read_csv("amazon_spending.csv")
df.head()
```

```python
# Full read of data
df.shape
```

```python
# Fill NaN with 0
df = df.fillna(0)
df.head()
```

```python
# Change item totals to floats from string
df["Item Total"] = df["Item Total"].str.replace('$','').astype(float)
df.head()
```

```python
# Commands for doing math with the data

# Total money spent on Amazon in the last year
#df['Item Total'].sum()

# Average amount spent per item in the last year
#df['Item Total'].mean()

# Most expensive item bought on Amazon
#df['Item Total'].max()

# Least expensive item 
#df['Item Total'].min()
```

```python
# Total money spent on Amazon in the last year
total=df['Item Total'].sum()
total = round(total,2)
print('I spent $'+str(total)+' on Amazon over the last year.')

# Average amount spent per item in the last year
average=df['Item Total'].mean()
average = round(average,2)
print('I spent an average of $'+str(average)+' per item on Amazon over the last year.')

# Most expensive item bought on Amazon
maximum=df['Item Total'].max()
print('The most expensive thing I bought on Amazon in the last year was $'+str(maximum)+'.')

# Least expensive item 
minimum=df['Item Total'].min()
print('The least expensive thing I bought on Amazon in the last year was $'+str(minimum)+'.')
```

```python
# Calculate average sales tax percentage

# Step 1: Change sales tax to float
df["Item Subtotal Tax"] = df["Item Subtotal Tax"].str.replace('$','').astype(float)

# Step 2: Calculate sums
sum_total = df['Item Total'].sum()
sum_tax = df['Item Subtotal Tax'].sum()

# Step 3: Divide
avg_tax = sum_tax/sum_total

# Step 4: Change to percent & print
per_tax = avg_tax * 100
print('The average sales tax was '+ str(round(per_tax,2))+'%.')
```

```python
# Convert to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
# Import matplotlib for graph
import matplotlib
```

```python
# Calculate Daily Orders
daily_orders = df.groupby('Order Date').sum()['Item Total']
```

```python
daily_orders.plot.bar(rot=90, figsize=(20,10), color='thistle')
```

```python
daily_orders.plot(rot=90, figsize=(20,10), color='cadetblue', marker='*')
print('As you can see, my spending spiked in November as a result of holiday shopping.')
```
